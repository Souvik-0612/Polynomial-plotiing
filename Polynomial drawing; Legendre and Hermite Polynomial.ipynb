# Legendre Polynomial

The recursive relation is 
$$ nP_n(x) = (2n-1)xP_{n-1}(x) -(n-1)P_{n-2}(x)$$
Base coditions:
$$ P_0(x) = 1  $$
$$P_1(x) = x$$

```python
# Recursive relation for Legendre Polynomial
def P(n, x): 
    if(n == 0):
        return 1 
    elif(n == 1):
        return x 
    else:
        return ((2*n - 1)*x*P(n-1, x)-(n-1)*P(n-2, x))/n
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)
plt.plot(x, np.ones((100)), label = "P0")
for i in range(1,5):
    plt.plot(x, P(i,x), label = 'P'+str(i))  
    
plt.legend(loc = 'best')
plt.xlabel("$x$")
plt.ylabel("$P_n(x)$")
plt.grid()
plt.show() 
plt.show()
```

# Hermite Polynomial
$$ H_{n}(x) = 2xH_{n-1}(x) - 2nH_{n-2}(x)$$
Base conditions:
$$ H_{0}(x) = 1$$
$$ H_{1}(x) = 2x$$

```python
# Recursive relation for Hermite Polynomial
def H(n,x):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        return 2*x*H(n-1, x) - 2*n*H(n-2, x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)
plt.plot(x, np.ones((100)), label = "H0")
for i in range(1,4):
    plt.plot(x, H(i,x), label = 'H'+str(i))  
    
plt.legend(loc = 'best')
plt.xlabel("$x$")
plt.ylabel("$H_n(x)$")
plt.grid() 
plt.show()
```
